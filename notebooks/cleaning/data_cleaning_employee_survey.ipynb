```python
import pandas as pd
```

```python
hr_employee_survey_data = pd.read_csv('../../data/raw/hr_employee_survey.csv')
hr_employee_survey_data
```

```python
# check the data is null
data_null = hr_employee_survey_data.isnull().sum()
data_null
```

```python
# check the data is na
data_na = hr_employee_survey_data.isna().sum()
data_na
```

```python
# check the data types
hr_employee_survey_data.dtypes
```

```python
# rename the columns
hr_employee_survey_data = hr_employee_survey_data.rename(columns={
    'EmployeeID': 'employee_id',
    'Department': 'department',
    'WorkLifeBalance': 'work_life_balance',
    'ManagementSupport': 'management_support',
    'CareerGrowth': 'career_growth',
    'OverallSatisfaction': 'overall_satisfaction'
})

hr_employee_survey_data
```

```python
# export the data to csv
df = pd.DataFrame(hr_employee_survey_data)
df.to_csv('../../data/processed/hr_employee_survey.csv', index=False)
print("Data saved to ../../data/processed/hr_employee_survey.csv")
```
