```python
import  pandas as pd
```

```python
data_survey = pd.read_csv('../data/processed/hr_employee_survey.csv')
data_survey
```

```python
data_survey_groued = data_survey.groupby(['department']).agg({'overall_satisfaction': ['mean']}).reset_index()
data_survey_groued.columns = ['department', 'mean']
data_survey_groued = data_survey_groued.sort_values(by=['mean'], ascending=True)
data_survey_groued
```

```python
# calculate correlation between features
correlation = data_survey[['work_life_balance', 'management_support', 'career_growth', 'overall_satisfaction']].corr()

# correlation vs overall_satisfaction
correlation['overall_satisfaction'].sort_values(ascending=False)
```

```python
# calculate mean of work_life_balance by department
data_survey_wark_life_balance = data_survey.groupby(['department']).agg({'work_life_balance': ['mean']}).reset_index()
data_survey_wark_life_balance.columns = ['department', 'mean']
data_survey_wark_life_balance = data_survey_wark_life_balance.sort_values(by=['mean'], ascending=True)
data_survey_wark_life_balance
```

```python
# ตรวจสอบค่าสหสัมพันธ์
correlation = data_survey['career_growth'].corr(data_survey['overall_satisfaction'])
print(f"Correlation between Career Growth and Overall Satisfaction: {correlation:.2f}")
```
